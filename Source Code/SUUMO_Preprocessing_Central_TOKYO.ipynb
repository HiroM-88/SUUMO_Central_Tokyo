## price column

```python
# Check the unique values in price column
df["price"].unique()
```

```python
# Delete "※権利金含む"
df["price"] = df["price"].str.replace("※権利金含む", "")


# Create "億円" and "万円" column
df["price"] = df["price"].str.replace("円", "")
df["Oku"] = df["price"].str.extract(r"(\d+)(?=億)").fillna(0).astype("int64") # why int64 : integer overflow when using int32
df["Man"] = df["price"].str.extract(r"(\d+)(?=万)").fillna(0).astype("int64") # why int64 : integer overflow when using int32
df["total Price"] = df["Oku"]*100000000 + df["Man"]*10000

# df.drop(columns=["Oku", "Man", "price"], inplace=True)
df.head()
```

## address column

```python
# Check the unique values in address column
df["address"].unique()
```

```python
# Split address into three columns
df[["prefecture", "city", "town"]] = df["address"].apply(lambda x: re.split("[都区]",x)).apply(pd.Series)
df.head()
```

## station distance column

```python
# Split station_distance column into four parts
df["line"] = df["station_distance"].apply(lambda x: x.split("「")[0])

df["station"] = df["station_distance"].apply(lambda x: x.split("「")[1].split("」")[0])

df["way"] = df["station_distance"].apply(lambda x: re.split(r'[0-9]+', x.split('」')[1])[0])

df["walking time"] = df["station_distance"].apply(lambda x: x.split("歩")[1].split("分")[0]).astype(int)

df.head()
```

## area column

```python
# Remove m in area column
df["area [m2]"] = df["area"].apply(lambda x: x.split("m")[0]).astype(float)
df
```

## floor plan column

```python
# Check the unique values in floor_plan column
df["floor_plan"].unique()
```

```python
# Function for identifying L,D,K,S
def ldk_check(data):
    return {
      'L':1 if 'L' in data else 0,
      'D':1 if 'D' in data else 0,
      'K':1 if 'K' in data else 0,
      'S':1 if 'S' in data else 0
  }
```

```python
# Adjust values
df["floor_plan"] = df["floor_plan"].replace("1ワンルーム", "ワンルーム")
df["floor_plan"] = df["floor_plan"].replace(r"ワンルーム.*", "1", regex=True)
df["room"] = df["floor_plan"].str.extract('(\d+)')

# Apply defined function
ldks = df["floor_plan"].apply(ldk_check).apply(pd.Series)
df = pd.concat([df, ldks], axis=1)
df
```

## balcony column

```python
# Check the unique values in balcony column
df["balcony"].unique()
```

```python
# Remove m in balcony column
df["balcony [m2]"] = df["balcony"].apply(lambda x: x.split("m")[0])
df
```

## year column

```python
df["year"].unique()
```

```python
# Split year column into two parts
df["date_year"] = df["year"].apply(lambda x: x.split("年")[0])
df["date_month"] = df["year"].apply(lambda x: x.split("年")[1].split("月")[0])
df
```

```python
# Convert to datetime type and compute the age of building
now = datetime.datetime.now()
now_year = now.year
now_month = now.month

df['year'] = df['year'].str.replace('年', '-').str.replace('月', '')

df['year'] = pd.to_datetime(df['year'], format='%Y-%m')
# df["diff_month"] = (now_year-df["year"].dt.year)*12 + (now_month-df["year"].dt.month)
df["age"] = (now_year-df["year"].dt.year)
```
