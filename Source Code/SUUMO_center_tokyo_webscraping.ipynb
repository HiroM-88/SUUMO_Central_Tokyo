```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# List for storing collected data
data = []

# Initial URL (starting page)
url = "/jj/bukken/ichiran/JJ010FJ001/?ar=030&bs=011&ta=13&jspIdFlg=patternShikugun&sc=13101&sc=13102&sc=13103&sc=13104&sc=13105&sc=13113&kb=1&kt=9999999&mb=0&mt=9999999&ekTjCd=&ekTjNm=&tj=0&cnb=0&cn=9999999"
base_url = "https://suumo.jp"

# Current page number
page_num = 0


# Loop until there is no next page (e.g., when "次へ" is not found)
while url:
    # Full URL
    complete_url = base_url + url
    # Add 1
    page_num += 1
    # Output current scraping page
    print(f"Now Scraping page {page_num}: {complete_url}")

    # Get html using Beautiful Soup
    result = requests.get(complete_url)
    soup = BeautifulSoup(result.content, 'html.parser')

    # Find all relevant apartment information
    all_info = soup.find_all("div", class_="dottable dottable--cassette")

    # Scrape all apartment information
    for info in all_info:
        apt_info = {}

        # Extract apartment name
        apt_name = info.find("dd", class_="dottable-vm")
        apt_info["apt_name"] = apt_name.text if apt_name else "N/A"

        # Extract price
        price = info.find("span", class_="dottable-value")
        apt_info["price"] = price.text if price else "N/A"

        # Extract address and other details
        address = info.find_all("div", class_="dottable-line")[2]
        apt_info["address"] = address.find("dd").text if address else "N/A"
        apt_info["station_distance"] = address.find_all("dd")[1].text if len(address.find_all("dd")) > 1 else "N/A"

        # Extract area
        area = info.find("table", class_="dottable-fix")
        area_size = area.find_all("dd")[0]
        apt_info["area"] = area_size.text if area_size else "N/A"

        # Extract floor plan
        floor_plan = area.find_all("dd")[1]
        apt_info["floor_plan"] = floor_plan.text if floor_plan else "N/A"

        # Extract balcony size
        balcony = info.find_all("table", class_="dottable-fix")[1]
        balcony_size = balcony.find_all("dd")[0]
        apt_info["balcony"] = balcony_size.text if balcony_size else "N/A"

        # Extract age of building
        year = balcony.find_all("dd")[1]
        apt_info["year"] = year.text if year else "N/A"

        # Append the information to the data list
        data.append(apt_info)

    # Find the "Next" link (the next page)
    next_link = soup.find('a', text='次へ')

    # If the "Next" link exists, update the URL; otherwise, break the loop
    if next_link:
        url = next_link['href']
    else:
        url = None  # Stop when there is no next page

    # Sleep for a short period to avoid hitting the server too quickly
    time.sleep(1)
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_csv('suumo_central_Tokyo.csv', sep = ',',encoding='utf-8-sig', index=False)
```
